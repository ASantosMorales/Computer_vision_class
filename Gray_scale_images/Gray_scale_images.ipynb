# Weighted transformation of a colored image into gray-scale-image

There are many ways to transform a colored image into a gray-scale-image.

For this section, we will use the NTSC (*National Television System Comitee*) standar to get the *luma* value, i.e. the data component of luminosity of a digital image.

\begin{equation*}
luma = 0.29894*R + 0.58704*G + 0.11402*B
\end{equation*}

where:
- $R$ is the red-layer pixel value
- $G$ is the green-layer pixel value
- $B$ is the blue-layer pixel value

<br>
<br>

### Let's code!

First, we need to call the needed libreries and get the initial variables (images, dimensions, etc).

```python
# Import the needed libreries
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Getting the image
img = cv2.imread('Kitty.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Matplotlib expects images in RGB format but OpenCV provides them in BGR

# Show the current image
plt.imshow(img)
plt.show()

# Getting dimensions
dimensions = img.shape

print("This image has " + str(dimensions[0]) + " pixels in height, " + str(dimensions[1]) + " pixels in width and " + str(dimensions[2]) + " pixels in depth")

# Splitting the RGB components
img_red_layer = img[:, :, 0]
img_green_layer = img[:, :, 1]
img_blue_layer = img[:, :, 2]
```

<br>

Now, we need to operate each pixel of the original image to the the *luma-image*.

```python
gray_image = np.zeros((dimensions[0], dimensions[1]), np.uint8) # Create the template image. At the begin, it is a zero matrix

# Each pixel has to be tranformed according with the NTSC equation
for height in range(dimensions[0]):                                              # Run through all vertical pixels
    for width in range(dimensions[1]):                                           # Run through all horizontal pixels
        gray_image[height, width] = 0.29824*img_red_layer[height, width] + \
                                    0.58704*img_green_layer[height, width] + \
                                    0.11402*img_blue_layer[height, width]

plt.imshow(gray_image, cmap = 'gray')
plt.show()
```
